```python
import os
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

# Dossier contenant les fichiers parquet
processed_dir = '../data/processed'

# Liste tous les fichiers parquet dans le dossier
parquet_files = [f for f in os.listdir(processed_dir) if f.endswith('.parquet')]

# Lit chaque fichier parquet dans un DataFrame et les stocke dans un dictionnaire
dfs = {}
for file in parquet_files:
    file_path = os.path.join(processed_dir, file)
    df_name = os.path.splitext(file)[0]
    dfs[df_name] = pd.read_parquet(file_path)
    print(dfs[df_name])
```
